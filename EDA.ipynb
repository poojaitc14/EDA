## import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## loading dataset

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.shape
df.columns
df.info()
df.describe()
```

```python
df.dtypes
```

## Univariate analysis

```python
sns.histplot(df["Age"], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
for c in ["Sex", "Pclass", "Embarked"]:
    plt.figure()
    sns.countplot(x=c, data=df, order=df[c].value_counts().index)
    plt.title(f"Counts of {c}")
    plt.xticks(rotation=20)
    plt.show()
```

## Bivariate analysis

```python
target = "Survived"
df[target].value_counts(normalize=True)
```

```python
sns.barplot(x="Survived", y="Age", data=df)
plt.title("Age vs Survival Barplot")
plt.show()
```

```python
# Numeric feature vs target
for c in ["Age", "Fare"]:
    plt.figure()
    sns.boxplot(x=target, y=c, data=df)
    plt.title(f"{c} by survival outcome")
    plt.show()
```

## Multivariate analysis

```python
sns.catplot(x="Survived", y="Age", hue="Pclass", kind="bar", data=df)
plt.title('Age vs Survived by Class')
plt.show()
```

```python
sns.pairplot(df[['Survived','Age','Fare','Pclass']])
plt.show()
```

## Outliers

```python
for c in ["Age", "Fare"]:
    plt.figure()
    sns.boxplot(y=df[c])
    plt.title(f"Boxplot of {c}")
    plt.show()
```

```python
# Scatter plot to show bivariate outliers
plt.figure()
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=df, alpha=0.7)
plt.title("Fare vs Age (colored by survival)")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.show()
```

```python
# Comparing Survival Rates with a Violin Plot
sns.violinplot(data=df, x='Survived', y='Age', hue='Sex', split=True)
plt.xlabel('Survived')
plt.ylabel('Age')
plt.title('Survival Rates by Age and Gender')
plt.show()
```

```python
# Analyzing Trends with a Line Plot
sns.lineplot(x='Age', y='Fare', hue='Pclass', data=df)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs. Fare by Passenger Class')
plt.show()
```
